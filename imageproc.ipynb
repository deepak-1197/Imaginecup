```
numpy_image=Image.fromarray(numpy_array)
plt.imshow(numpy_image)
```

```
from PIL import ImageEnhance
plt.figure(figsize=(10,10))
image_color_enhan=image.copy()
image1=ImageEnhance.Color(image_color_enhan).enhance(2.5)
image2=ImageEnhance.Contrast(image_color_enhan).enhance(5.5)
image3=ImageEnhance.Brightness(image_color_enhan).enhance(1.5)
image4=ImageEnhance.Sharpness(image_color_enhan).enhance(3.5)
plt.subplot(2,2,1)
plt.imshow(image1)
plt.subplot(2,2,2)
plt.imshow(image2)
plt.subplot(2,2,3)
plt.imshow(image3)
plt.subplot(2,2,4)
plt.imshow(image4)

```

```
image_blend=Image.blend(image1,image4,0.5)
plt.imshow(image_blend)
```

```
img_channel=image.copy()
r,g,b=img_channel.split()
im=Image.merge("RGB",(r,b,g))
plt.imshow(im)
```

```
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
img=cv2.imread("emma.jpg")
kernel=np.ones((3,3),np.uint8)
cv2_imshow(img)
```

```
imgGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgBlur=cv2.GaussianBlur(img,(5,5),0)
imgCanny=cv2.Canny(img,150,200)
imgDialation=cv2.dilate(imgCanny,kernel,iterations=1)
#cv2_imshow(imgGray)
#cv2_imshow(imgBlur)
cv2_imshow(imgCanny)
cv2_imshow(imgDialation)
```

```
imga=np.zeros((512,512,3),np.uint8)
cv2.line(imga,(0,0),(imga.shape[1],imga.shape[0]),(0,255,0),3)
cv2.rectangle(imga,(250,250),(500,imga.shape[1]-100),(0,0,255),2)
cv2.circle(imga,(400,50),30,(255,255,0),5)
cv2_imshow(imga)
```
